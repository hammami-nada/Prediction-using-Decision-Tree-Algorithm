# Task 1:Prediction using Decision Tree Algorithm

In this task,we will create a Decision Tree clasifier and visualize it grafically. 

The purpose is if we feed any new data to this classifier it would be able to predict the right class accordingly based on its learned patterns and features from the training data.
In essence, the purpose of the classifier is to accurately classify new, unseen data into the appropriate categories based on its learned knowledge from the training phase.

```python
# Importing libraries 
# The Iris dataset is readily available through scikit-learn's datasets module, so there's no need to load it from an external CSV file
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
# loading the iris dataset
iris=datasets.load_iris()
x=iris.data
y=iris.target # will contain the species of iris corresponding to each sample, encoded as integers
```

```python
# Instantiate DecisionTreeClassifier
dec = DecisionTreeClassifier()

# Fit the model to your data
model = dec.fit(x, y)
```

```python
# Visualising tree with plot_tree
fig=plt.figure(figsize=(25,20))
# "_ "is often used as a convention to indicate that the return value of the function call is not going to be used explicitly in the code
_=tree.plot_tree(dec,
              feature_names=iris.feature_names, # that contains the names of the features(sepal length, sepal width, petal length, petal width) 
              class_names=iris.target_names, # a property of the Iris dataset that contains the names of the target classes (e.g., Setosa, Versicolor, Virginica)
              filled=True) #  the decision tree will be drawn with filled boxes representing each node
```

You can now feed any new/test data to this classifer and it would be able to predict the right class accordingly.
